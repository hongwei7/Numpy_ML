```python
import pandas as pd
file=pd.read_csv('mnist_train.csv')
print(file.shape)
print(file)
mnist=dict()
print(file.columns)
mnist['data']=file
mnist['target']=file['label']
del mnist['data']['label']
mnist['data']=mnist['data'].values
mnist['target']=mnist['target'].values
X,y=mnist['data'],mnist['target']
X_train,y_train,X_test,y_test=X[:-100],y[:-100],X[:100],y[:100]
import numpy as np
shuffle_index=np.random.permutation(50000)
X_train,y_train=X_train[shuffle_index],y_train[shuffle_index]
print('done')
```

```python
X_train=np.matrix(X_train)
X_test=np.matrix(X_test)
y_train=np.matrix(y_train).T
y_test=np.matrix(y_test).T
```

```python
class KNN(object):
    def __init__(self,X,y):
        self.X=X
        self.y=y
    def single_predict(self,x):
        diffmat=np.repeat(x,self.X.shape[0]).reshape(x.shape[1],self.X.shape[0]).T-self.X
        squarediffmat=(np.matrix(np.array(diffmat)**2)).sum(axis=1)
        return list(self.y[np.where(squarediffmat==np.min(squarediffmat))[0]][0])[0]
    def predict(self,x):
        result=[]
        k=0
        for i in x:
            k+=1
            print(k)
            result.append(self.single_predict(i))
        return result
clf=KNN(X_train,y_train)
pre_y=clf.predict(X_test)
```

```python
s,k=0,0
for i,j in zip(pre_y,y_test):
    k+=1
    s+=(int(i==j))
s/k*100,'%'
```
